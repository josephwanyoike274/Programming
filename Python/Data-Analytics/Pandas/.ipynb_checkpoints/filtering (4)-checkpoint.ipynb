```python
import pandas as pd
```

```python
people = {
    "first_name" : ["joseph", "monicah", "lydia","monicah"],
    "last_name": ["wanyoike", "wambui", "mwaura","wanyoike"],
    "email" : ["njw@gmail.com", "mwg@gmail.com", "lwm@gmail.com", "mw@gmail.com"]
    
}

```

```python
df_people = pd.DataFrame(people)
```

```python
df_people
```

```python
df_people["last_name"] == "wanyoike"
```

```python
filt = (df_people ["last_name"] == "wanyoike")
```

```python
df_people [filt]
```

```python
df_people[df_people ["last_name"] == "wanyoike"]#works same way as above
```

```python
#lets now use loc
```

```python
df_people.loc[filt]
```

```python
df_people.loc[filt, "email"]#advantage of using loc is that we can even specify which columns we want 
```

```python
#and or or keywords & and |
```

```python
filt2 = (df_people["last_name"] == "wanyoike") &  (df_people["first_name"] == "monicah")
```

```python
filt2
```

```python
df_people.loc[filt2]#display the rows that meets those two conditions
```

```python
filt3 = (df_people["last_name"] == "wanyoike") |  (df_people["first_name"] == "monicah")#any of the conditions
```

```python
df_people.loc[filt3]
```

```python
df_people[-filt3]#adding - (negative sign) negates the statement , it display rows that dont meet that criteria
```
